```python
import numpy as np

# XOR dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([
    [0],
    [1],
    [1],
    [0]
])

# Sigmoid activation
def sigmoid(x):
    return 1/(1 + np.exp(-x))

# Derivative of sigmoid
def sigmoid_derivative(x):
    return x*(1-x)

np.random.seed(0)

# Weights initialization
W1 = np.random.uniform(-1,1,(2,4))   # input → hidden
b1 = np.zeros((1,4))

W2 = np.random.uniform(-1,1,(4,1))   # hidden → output
b2 = np.zeros((1,1))

learning_rate = 0.5
epochs = 20000

for epoch in range(epochs):

    # Forward pass
    hidden_input = np.dot(X, W1) + b1
    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, W2) + b2
    final_output = sigmoid(final_input)

    # Error
    error = y - final_output

    # Backpropagation
    d_output = error * sigmoid_derivative(final_output)

    error_hidden = np.dot(d_output, W2.T)
    d_hidden = error_hidden * sigmoid_derivative(hidden_output)

    # Update weights
    W2 += np.dot(hidden_output.T, d_output) * learning_rate
    b2 += np.sum(d_output, axis=0, keepdims=True) * learning_rate

    W1 += np.dot(X.T, d_hidden) * learning_rate
    b1 += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate

    # Print loss
    if epoch % 5000 == 0:
        loss = np.mean(np.square(error))
        print(f"Epoch {epoch}, Loss: {loss}")

# Final predictions
print("\nFinal Predictions:")
print(final_output)
```

```python
import numpy as np

# XOR dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([
    [0],
    [1],
    [1],
    [0]
])

# ReLU activation
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

np.random.seed(0)

# Weights initialization
W1 = np.random.uniform(-1,1,(2,4))   # input → hidden
b1 = np.zeros((1,4))

W2 = np.random.uniform(-1,1,(4,1))   # hidden → output
b2 = np.zeros((1,1))

learning_rate = 0.01
epochs = 20000

for epoch in range(epochs):

    # Forward pass
    hidden_input = np.dot(X, W1) + b1
    hidden_output = relu(hidden_input)   # ReLU hidden

    final_input = np.dot(hidden_output, W2) + b2
    final_output = relu(final_input)     # ReLU output

    # Error
    error = y - final_output

    # Backpropagation
    d_output = error * relu_derivative(final_output)

    error_hidden = np.dot(d_output, W2.T)
    d_hidden = error_hidden * relu_derivative(hidden_output)

    # Update weights
    W2 += np.dot(hidden_output.T, d_output) * learning_rate
    b2 += np.sum(d_output, axis=0, keepdims=True) * learning_rate

    W1 += np.dot(X.T, d_hidden) * learning_rate
    b1 += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate

    # Print loss
    if epoch % 1000 == 0:
        loss = np.mean(np.square(error))
        print(f"Epoch {epoch}, Loss: {loss}")

# Final predictions (apply threshold)
print("\nFinal Predictions (thresholded):")
print((final_output > 0.5).astype(int))
```

```python
import numpy as np

# XOR dataset (one-hot labels for 2 classes)
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([
    [1,0],  # class 0
    [0,1],  # class 1
    [0,1],  # class 1
    [1,0]   # class 0
])

# Tanh activation and derivative
def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

# Mean squared error loss
def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

np.random.seed(0)

# Weights initialization
W1 = np.random.uniform(-1,1,(2,4))   # input → hidden
b1 = np.zeros((1,4))

W2 = np.random.uniform(-1,1,(4,2))   # hidden → output (2 classes)
b2 = np.zeros((1,2))

learning_rate = 0.1
epochs = 20000

for epoch in range(epochs):
    # Forward pass
    hidden_input = np.dot(X, W1) + b1
    hidden_output = tanh(hidden_input)   # <-- tanh activation

    final_input = np.dot(hidden_output, W2) + b2
    final_output = final_input           # <-- linear output (no softmax)

    # Loss
    loss = mse(y, final_output)

    # Backpropagation
    d_output = (final_output - y)        # derivative of MSE wrt output
    error_hidden = np.dot(d_output, W2.T)
    d_hidden = error_hidden * tanh_derivative(hidden_input)

    # Update weights
    W2 -= np.dot(hidden_output.T, d_output) * learning_rate
    b2 -= np.sum(d_output, axis=0, keepdims=True) * learning_rate

    W1 -= np.dot(X.T, d_hidden) * learning_rate
    b1 -= np.sum(d_hidden, axis=0, keepdims=True) * learning_rate

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {loss}")

# Final predictions
print("\nFinal Predictions:")
print(np.argmax(final_output, axis=1))
```
